```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.random.randint(10,size=5)
y=np.random.randint(10,size=5)
ax = sns.barplot(x,y)
ax.set(xlabel='X-axis', ylabel='Height')
```

```python
x=np.random.random(100)*100
ax = sns.distplot(x, bins=np.arange(0,101,10))
ax.set(xlabel='Element', ylabel='Density')
ax.set(title='Distplot')
```

```python
x=np.random.randint(10,size=10)
y=np.random.randint(10,size=10)
hue=np.random.randint(2,size=10)
ax=sns.scatterplot(x,y,hue=hue)
ax.set(xlabel='X-Axis', ylabel='Y-Axis')
ax.set(title='Scatter Plot')
```

```python
x=np.random.random(100)
y=np.random.random(100)*2
hue=np.random.randint(2,size=100)
df=pd.DataFrame()
df['x']=x
df['y']=y
df['Hue']=hue
sns.lmplot('x', 'y',data=df, hue='Hue')
```

```python
x=np.random.random(100)*100
sns.kdeplot(x)
plt.xlabel('Elements')
plt.ylabel('Density')
plt.title('KDE Plot')
```

```python
x=np.arange(20)
y=np.random.random_integers(20,size=20)
print(x)
print(y)
ax=sns.lineplot(x,y)
ax.set(xlabel='X-axis',ylabel='Y-axis')
ax.set(title='Lineplot')
```

```python
x=range(1,6)
y=np.random.randint(5,size=50)
ax = sns.distplot(y,bins=x,kde=False)
ax.set(xlabel='Bins',ylabel='Frequency')
ax.set(title='Histogram')
```

```python
x=np.random.randint(50,size=50)
x=np.append(x,100)
ax=sns.boxplot(x)
ax.set(xlabel='Range',ylabel='Y-Axis')
ax.set(title='Boxplot')
```

```python
iris = sns.load_dataset("iris")
sns.pairplot(iris)
```

```python
x=np.random.randint(2,size=20)
y=np.random.randint(10,size=20)
ax=sns.violinplot(x,y)
ax.set(xlabel='Elements',ylabel='Density')
ax.set(title='Violinplot')
```

```python
iris = sns.load_dataset("iris")
ax=sns.stripplot(data=iris)
ax.set(title='Strip Plot')
```

```python
iris = sns.load_dataset("iris")
ax=sns.swarmplot(data=iris)
ax.set(title='Swarm Plot')
```
